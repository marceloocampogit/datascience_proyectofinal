**5.1 Selecciono del DataSet las variables a considerar para entrenar mi modelo**

```python
df = df[['TIPO', 'PORCENTAJE AUTO', 'MARCA', 'VALOR ACTUAL DEL AUTO', 'PROVINCIA', 'ESTADO CIVIL', 'SITUACION LABORAL', 'SCORE CLIENTE', 'RIESGO']]
```

**5.2 Transformamos variables categóricas a numéricas.**

```python
df = pd.get_dummies(df)
df.head()
```

```python
X = df.drop("RIESGO", axis=1) #Elimino de mi dataset la variable a predecir
y = df.RIESGO #Defino el Target
```

**5.3 Hacemos la división entre la data de evalución y de entrenamiento**

```python
#Importamos la libreria para separar en train y test
from sklearn.model_selection import train_test_split 

#Separamos variables de train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=440)
```

**5.4 Decision Tree Classifier**




**5.4.1 Impletación del modelo**

```python
from sklearn.tree import DecisionTreeClassifier #Importamos el modelo
tree = DecisionTreeClassifier(max_depth= 3, random_state = 440) #Creamos el modelo
```

**5.4.2 Entrenamiento del modelo**

```python
tree.fit(X_train,y_train)
```

**5.4.3 Predicción sobre Train y Test**

```python
y_train_pred = tree.predict(X_train) #Prediccion en Train
y_test_pred = tree.predict(X_test) #Prediccion en Test
```

**5.4.4 Evaluamos el % de acierto en ambos lados: Accuracy**


```python
from sklearn.metrics import accuracy_score

#Calculo el accuracy en Train
train_accuracy_dt = accuracy_score(y_train, y_train_pred)

#Calculo el accuracy en Test
test_accuracy_dt = accuracy_score(y_test, y_test_pred)

print('% de aciertos sobre el set de entrenamiento:', train_accuracy_dt)
print('% de aciertos sobre el set de evaluación:',test_accuracy_dt)
```

**5.4.5 Evaluamos el % de acierto en ambos lados: Precision**

```python
from sklearn.metrics import precision_score

#Calculo la precision en Train
train_precision_dt = precision_score(y_train, y_train_pred, average='weighted', zero_division= 0)

#Calculo la precision en Test
test_precision_dt = precision_score(y_test, y_test_pred, average='weighted', zero_division= 0)

print('% de aciertos sobre el set de entrenamiento:', train_precision_dt)
print('% de aciertos sobre el set de evaluación:',test_precision_dt)
```

**5.4.6 Evaluamos el % de acierto en ambos lados: Sensibilidad**

```python
from sklearn.metrics import recall_score

#Calculo la precision en Train
train_recall_dt = recall_score(y_train, y_train_pred, average='weighted')

#Calculo la precision en Test
test_recall_dt = recall_score(y_test, y_test_pred, average='weighted')

print('% de aciertos sobre el set de entrenamiento:', train_recall_dt)
print('% de aciertos sobre el set de evaluación:',test_recall_dt)
```

**5.4.7 Evaluamos el % de acierto en ambos lados: F1 SCORE**

```python
from sklearn.metrics import f1_score

#Calculo la precision en Train
train_f1_dt = f1_score(y_train, y_train_pred, average='weighted')

#Calculo la precision en Test
test_f1_dt = f1_score(y_test, y_test_pred, average='weighted')

print('% de aciertos sobre el set de entrenamiento:', train_f1_dt)
print('% de aciertos sobre el set de evaluación:',test_f1_dt)
```
